# **CHAIN WITH CUSTOM RUNNABLE**

```python
# Task 1 - Prompt

from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are an helpful assistant"),
    ("human", "{input}")
])
```

```python
# Task 2 - LLM

llm_groq = init_chat_model(model = "llama-3.1-8b-instant", model_provider="groq")
```

```python
# Task 3 - String Parser

str_parser = StrOutputParser()
```

```python
# Task 4 - Custom Runnable

from langchain_core.runnables import RunnableLambda

def dictionary_maker(text:str)->dict:

    return {"text" : text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)
```

```python
# Task 5 - Template For Post

prompt_post = ChatPromptTemplate.from_messages([
    ("system","You are a social media post generator"),
    ("human","Create a post for the following text for Instagram: {text}")
])
```

```python
# Task 6 - Generate the Post

llm_groq = init_chat_model(model = "llama-3.1-8b-instant", model_provider="groq")
```

```python
# Task 7 - Output Parser

str_parser = StrOutputParser()
```

```python
# Chain

chain = prompt_template | llm_groq | str_parser | dictionary_maker_runnable | prompt_post | llm_groq | str_parser

chain.invoke("What is the capital of Karnataka?")
```
