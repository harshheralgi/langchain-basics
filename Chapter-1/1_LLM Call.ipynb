# **LLM CALL**

```python
#Traditional Way!

import os
from groq import Groq

client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Hello!"}
    ]
)

print(response.choices[0].message.content)
```

```python
#Using LangChain!

from dotenv import load_dotenv
import os
from load_dotenv import load_dotenv
from langchain_groq import ChatGroq
 
load_dotenv()

llm_groq = ChatGroq(model = "llama-3.1-8b-instant", temperature = 0)

llm_groq.invoke("Tell a joke")
```

```python
#Using Chat Models Init Feature!

from langchain.chat_models import init_chat_model

llm_groq = init_chat_model(model = "llama-3.1-8b-instant", model_provider="groq")

llm_groq.invoke("Hello Bro, Harsh Here! This is a check for chat init model in Langchain!")
```
