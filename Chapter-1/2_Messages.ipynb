# **MESSAGES**

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain.chat_models import init_chat_model
import os
from load_dotenv import load_dotenv

load_dotenv()

llm_groq = init_chat_model(model = "llama-3.1-8b-instant", model_provider="groq")

my_messages = [
    SystemMessage(content="You are a Gen-z Assistant who always answers in a fun way!"),
    HumanMessage(content="Broo! Tell me a joke")
]

llm_groq.invoke(my_messages).content
```

```python
my_messages = [
    SystemMessage(content="You are a Gen-z Assistant who always answers in a fun way!"),
    HumanMessage(content="Broo! Tell me a fun fact about Honey")
]

llm_groq.invoke(my_messages).content
```

# **PROMPTS**

```python
from langchain_core.prompts import PromptTemplate

user_input = input("Enter a topic for fun fact: ")

dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic" : user_input})

llm_groq.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

user_input = input("Enter a topic")

prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are an assistant who always roasts"),
    ("user","Write a fun fact about {topic}")
])

ready_prompt = prompt_template.invoke({"topic" : user_input})

llm_groq.invoke(ready_prompt).content
```
