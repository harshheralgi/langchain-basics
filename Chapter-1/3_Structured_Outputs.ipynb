```python
from langchain_groq import ChatGroq
import os
from load_dotenv import load_dotenv

load_dotenv()

if os.environ.get("GROQ_API_KEY"):
    print("API KEY FETCHED")
else:
    raise ValueError("GROQ_API_KEY Not Found!!")
```

```python
from langchain.chat_models import init_chat_model

llm_groq = init_chat_model(model = "llama-3.1-8b-instant", model_provider="groq")

llm_groq.invoke("Hello!")
```

```python
from langchain_core.prompts import PromptTemplate

result = llm_groq.invoke("Tell me a joke. Generate the output in the key value format with the following keys: setup, punchline")

result.content
```

# **Using Pydantic Models**

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    setup : str = Field(description="The setup of the joke")
    punchline : str = Field(description="The punchline of the joke")
```

```python
obj = llm_schema(**{"setup":"Some Setup" , "punchline":"Some Punchline"})
obj
```

```python
llm_structured_output = llm_groq.with_structured_output(llm_schema)

llm_structured_output.invoke("Tell me a joke!")
```

# **Using TypedDict**

```python
from typing import TypedDict

class llm_schema_td(TypedDict):
    setup :str
    punchline : str
```

```python
llm_structured_typed_dict = llm_groq.with_structured_output(llm_schema_td)

result = llm_structured_typed_dict.invoke("Tell me a joke!")

result
```
